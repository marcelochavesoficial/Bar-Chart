# **Bar Chart**

The bar chart is a chart made up of rectangular shaped bars. The length (or height) of each bar is proportional to the value it represents. It is worth mentioning that it is common to see both bars drawn vertically and horizontally.

No truth is absolute, but the bar graph is much more useful when trying to represent discrete data. Although many people confuse it with the histogram, they are completely different concepts.

# Plot the bar chart

The data we will use in this notebook are public data from the São Paulo subway.

Datasets containing various statistics can be found on the Corporate Governance and Transparency Portal. On this Portal, you can find Audit Committee deliberations, composition of the councils, and information from all areas of the subway, being able to follow financial management, new projects in gestation and the progress of works and their evolution.

Although they were made available in csv format, the dataset "Passenger Entry by Station - Average Working Days - December/2018" needed to be worked on and structured.

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# import data to dataframe
data = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/entrada_passageiro_linha1.csv"
df = pd.read_csv(data)

# view first 5 entries
df.head()
```

```python
# view the last 5 entries
df.tail()
```

```python
# delete line TOTAL
df.drop(index=23, axis=0, inplace=True)
```

```python
# plot the bar chart with the average passenger entry
fig, ax = plt.subplots(figsize=(10,6))

df.plot(x="Estação", y="Entradas", kind="bar", ax=ax)

ax.set_title("Average Passenger Entry - Line 1 (December/2018)")
ax.set_xlabel("Estação")
ax.set_ylabel("Entradas")

plt.tight_layout()
```
